```python
def pad_numpy_image_to_fit_patch(image, patch_size):
    batch_size, _, height, width = image.shape

    # calculate the number of pixels to add to width and height
    pad_height = np.ceil(height / patch_size) * patch_size - height
    pad_width = np.ceil(width / patch_size) * patch_size - width

    # pad the image
    padded_image = np.pad(
        image,
        pad_width=((0, 0), (0, 0), (0, int(pad_height)), (0, int(pad_width))),
        mode='constant',
        constant_values=0
    )

    return padded_image


class DinoV2Preprocessor(torch.nn.Module):
    MEAN = (0.485, 0.456, 0.406)
    STD = (0.229, 0.224, 0.225)

    @torch.no_grad()
    def forward(self, image: torch.Tensor) -> torch.Tensor:
        image = (image / 255.0).float()
        image = tvf.normalize(image, mean=list(self.MEAN), std=list(self.STD))
        return image
```
