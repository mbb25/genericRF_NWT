```python
#----------------------------------------------------------------------------------------------------
# Data pipeline parameters
#----------------------------------------------------------------------------------------------------

# DATA DIRECTORY
data_dir = 'C:/Users/MABEB16/science/postdoc/5_projects/1_general_classifier/2_labelling/hp' #MBB

# OUTPUT DIR and OUTPUT FILENAME
csv_file = 'dataset_file.csv'
outdir = 'C:/Users/MABEB16/science/postdoc/5_projects/1_general_classifier/4_general_classifier/results/' # MBB

# CLASSES
class_map = {
    'lichen': 1,
    'rock': 4,
    'broadleaf': 5,
    'needleleaf': 6,
    'deadwood': 7,
    'graminoids': 8,
    'moss': 9,
    'soil': 10,
    'low_green': 12,
    'dry_branches': 13,
}

# SITES
#sites = ['CG1-8A'] #MBB
sites = os.listdir(data_dir) # all sites
print(sites)
print(len(sites))
```

# Run data extraction
Run data extraction on all sites in list *sites*. Data is extracted for all labelled pixels from site-specific shape and raster files, then concatenated into one big dataframe (*all_features*).
<br>**NOTE:** File dirs and names are currently hardcoded in the load_site_data() function.</br>

### Required Files in the *data_dir/{site}* directory:
- {site}_hp_transparent_mosaic_group1.tif
- {site}_hp_chm_stratified.tif
- {site}_hp_chm_stratified.shp
- {site}_hp_labelledpoints.gpkg

```python
features = []
# sites = ['SS33-1C', 'SS33-4A', 'SS33-9A', 'ZF20-4A', 'ZF20-6B', 'ZF20-8B', 'ZF20-9A', 'ZF20-9C']
#sites = ['SS33-1C']
for site in sites:
    print(f'Processing {site}...')
    combined_raster = load_site_data(data_dir, site)
    df, _ = convert_to_dataframe(combined_raster, site, class_map=class_map)
    features.append(df)

# The all_features dataframe contains for all labeled pixels information on 
# site, y- and x-position, class label, RGB, chm and labelling certainty
all_features = pd.concat(features, axis=0)

# normalize RGB values in all_features dataframe
df = normalize_rgb(all_features)

# write to csv
df.to_csv(outdir+csv_file, index=False)
# all features
```
